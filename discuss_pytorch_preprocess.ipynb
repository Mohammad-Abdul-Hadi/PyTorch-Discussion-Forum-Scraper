```
import pandas as pd
import os
from bs4 import BeautifulSoup
```

```
def get_desc(html_code):
  if str(html_code)=="nan":
    return "nan"
  else:
    soup = BeautifulSoup(html_code)
  return soup.text


def get_replies(html_code):
  if str(html_code)=="nan":
    return 0, "nan"
  else:
    soup = BeautifulSoup(html_code)
    j=""
    length = len(soup.text.split("REPLY")[1:])
    for i in soup.text.split("REPLY")[1:length]:
      j= j + i.split(":",1)[1].strip()+"<SPLIT>"
    j = j + soup.text.split("REPLY")[1:][-1].split(":")[1].strip()
  return length, j

def get_repliers(html_code):
  if str(html_code)=="nan":
    return "nan"
  else:
    soup = BeautifulSoup(html_code)
    j=""
    length = len(soup.text.split("REPLIER")[1:])
    for i in soup.text.split("REPLIER")[1:length]:
      j= j + i.split(":",1)[1].strip().split("/")[-1].strip(";")+"<MRR-SPLIT>"
    j = j + soup.text.split("REPLIER")[1:][-1].split(":", 1)[1].strip().split("/")[-1].strip(";")
  return j

def get_reply_dates(html_code):
  if str(html_code)=="nan":
    return "nan"
  else:
    soup = BeautifulSoup(html_code)
    j=""
    length = len(soup.text.split("REPLY_DATE")[1:])
    for i in soup.text.split("REPLY_DATE")[1:length]:
      j= j + i.split(":",1)[1].strip().split("/")[-1].strip(";")+"<MRR-SPLIT>"
    j = j + soup.text.split("REPLY_DATE")[1:][-1].split(":",1)[1].strip().strip(";")
  return j

def get_reply_likes(html_code):
  if str(html_code)=="nan":
    return "nan"
  else:
    soup = BeautifulSoup(html_code)
    j=""
    length = len(soup.text.split("REPLY")[1:])
    for i in soup.text.split("REPLY")[1:length]:
      j= j + i.split(":",1)[1].strip().split("/")[-1].strip(";")+"<MRR-SPLIT>"
    j = j + soup.text.split("REPLY")[1:][-1].split(":",1)[1].strip().strip(";")
    j = j.replace(" Like","")
    arr = j.split("<MRR-SPLIT>")
    st = ""
    for j in arr:
      if j.strip()=="":
        st = st + "0" + "<MRR-SPLIT>"
      else:
        st = st + j + "<MRR-SPLIT>"

  return st

def check_code(html_code):
  if str(html_code)=="nan":
    return False, "nan"
  else:
    soup = BeautifulSoup(html_code)
    a = ""
    for link in soup.findAll('code'):
      a = a + link.text
    if a=="":
      return False, "nan"
    else:
      return True, a

def answered_by_ofic(reps, moderators):
  if reps=="nan":
    return "nan"
  repls = reps.split("<MRR-SPLIT>")
  for i in repls:
    if i in moderators:
      return True
    else:
      return False
```

```
!mkdir /content/PyTorch-Discussion-Forum-Scraper/processed-data
% cd /content/PyTorch-Discussion-Forum-Scraper/Dataset
all_csv = os.listdir()
all_csv.remove('Titles and Metadata for each Category')


for fle in all_csv:
  %cd /content/PyTorch-Discussion-Forum-Scraper/Dataset
  df = pd.read_csv(fle)
  all_desc = df["description"].values 
  all_replies = df["replies"].values
  all_repliers = df["repliers_links"].values 
  all_reply_dates = df["reply_dates"].values 
  all_reply_likes = df["reply_likes"].values
  fin_desc = []
  fin_replies = []
  no_comments = []
  fin_repliers = []
  fin_has_code = []
  fin_reply_dates =[]
  fin_reply_likes = []
  fin_code =[]
  fin_ofic_reply = []
  mods = ["ptrblck", "albanD", "fmassa" ,"jspisak","smth", "colesbury","apaszke", "richard","ezyang","InnovArul","SimonW","driazati","David_Reiss"]
  for row in range(len(all_desc)):

    fin_desc.append(get_desc(all_desc[row]))#
    length, j = get_replies(all_replies[row])#
    no_comments.append(length)#
    fin_replies.append(j)#
    fin_repliers.append(get_repliers(all_repliers[row]))#
    fin_reply_dates.append(get_reply_dates(all_reply_dates[row]))#
    fin_reply_likes.append(get_reply_likes(all_reply_likes[row]))#
    ha_code, cod = check_code(all_desc[row])#
    fin_has_code.append(ha_code)#
    fin_code.append(cod)#
    fin_ofic_reply.append(answered_by_ofic(get_repliers(all_repliers[row]), mods))#
  df["proc-description"]=fin_desc
  df["desc-has-code"] = fin_has_code
  df["desc-code"]= fin_code
  df["no-replies"] = no_comments
  df["proc_replies"] = fin_replies
  df["proc-repliers"] = fin_repliers
  df["reply-from-ofic"] = fin_ofic_reply
  df["proc-reply-dates"] = fin_reply_dates
  df["proc-reply-likes"] = fin_reply_likes
  % cd /content/PyTorch-Discussion-Forum-Scraper/processed-data
  df.to_csv("new-"+str(fle), index=False)
```
